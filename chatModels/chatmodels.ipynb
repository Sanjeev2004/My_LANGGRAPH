```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
import os

# Load .env
load_dotenv()

# Initialize
model = ChatOpenAI(model="gpt-4o-mini", temperature=0, max_tokens=10)

# Call the model
result = model.invoke("Hello! How are you?")
print(result.content)
```
